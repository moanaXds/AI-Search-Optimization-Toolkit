```python


import random

# -----------------------------
# Problem Data
# -----------------------------

items = ["N1", "N2", "N3", "N4", "N5", "N6"]

values = [14, 23, 8, 9, 17, 15]
weights = [1, 3, 7, 4, 5, 6]

capacity = 10

POP_SIZE = 10
MAX_GEN = 20
MUTATION_PROB = 0.1


# -----------------------------
# Create Initial Population
# -----------------------------

def initialize_population():
    population=[]

    for i in range (POP_SIZE):
        chromosomes=[]
        for i in range(6):
            chromosomes.append(random.randint(0,1))  # genes =0/1

        population.append(chromosomes)

    return population
# -----------------------------
# Fitness Function
# -----------------------------
def fitness(chromosome):

    total_value = 0
    total_weight = 0

    for i in range(len(chromosome)):
        if chromosome[i] == 1:
            total_value += values[i]
            total_weight += weights[i]

    if total_weight > capacity:
        return 0   # penalty

    return total_value


# -----------------------------
# Tournament Selection
# -----------------------------
def tournament_selection(population):

    k = 3
    candidates = random.sample(population, k)

    best = max(candidates, key=fitness)

    return best


# -----------------------------
# One Point Crossover
# -----------------------------
def crossover(parent1, parent2):

    point = random.randint(1, len(parent1) - 1)

    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]

    return child1, child2


# -----------------------------
# Mutation
# -----------------------------
def mutate(chromosome):

    for i in range(len(chromosome)):
        if random.random() < MUTATION_PROB:
            chromosome[i] = 1 - chromosome[i]

    return chromosome


# -----------------------------
# Replace Least Fit Individuals
# -----------------------------
def replacement(population, children):

    population = sorted(population, key=fitness)

    population[0] = children[0]
    population[1] = children[1]

    return population


# -----------------------------
# Main Genetic Algorithm
# -----------------------------
def genetic_algorithm():

    population = initialize_population()

    for i in range(MAX_GEN):

        parent1 = tournament_selection(population)
        parent2 = tournament_selection(population)

        child1, child2 = crossover(parent1, parent2)

        child1 = mutate(child1)
        child2 = mutate(child2)

        population = replacement(population, [child1, child2])

        best = max(population, key=fitness)
        print("Generation:", i, "Best:", best, "Fitness:", fitness(best))


# -----------------------------
# Run Program
# -----------------------------
if __name__ == "__main__":
    genetic_algorithm()


# In[ ]:
```
